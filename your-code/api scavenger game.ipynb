Challenge 2: Count Commits

```python
from requests.auth import HTTPBasicAuth
from getpass import getpass
from requests_oauthlib import OAuth2Session

token = getpass()
```

```python
url1 = "https://api.github.com/repos/janec4s/lab-list-comprehensions/commits"
response1 = requests.get(url = url1, auth = HTTPBasicAuth('janec4s', token))
response1.status_code
```

```python
commits = response1.json()
commits
```

```python
comm = pd.json_normalize(commits)
comm
```

```python
comm["node_id"].count()
```

```python
comm[["commit.author.date", "commit.committer.date"]]
```

```python
comm["commit.author.date"] = pd.to_datetime(comm["commit.author.date"])
```
