```python
def normalize_column(data, column):
  flat = pd.DataFrame(dict(data[column])).transpose()
  columns = [str(col) for col in flat.columns]
  data = pd.concat([data, flat], axis = 1)
  data = data.drop(column, axis = 1)
  return data

flat_df = normalize_column(data, 'commit')
flat_df.head()
```
