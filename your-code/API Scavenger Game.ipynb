### Challenge 1: Fork Languages

You will find out how many programming languages are used among all the forks created from the main lab repo of your bootcamp. Assuming the main lab repo is  `ironhack-datalabs/bcn-feb-2019`, you will:

1. Obtain the full list of forks created from the main lab repo via Github API.

1. Loop the JSON response to find out the `language` attribute of each fork. Use an array to store the `language` attributes of each fork.
    * *Hint: Each language should appear only once in your array.*

1. Print the language array. It should be something like:

	```["Python", "Jupyter Notebook", "HTML"]```

Again, the documentation of Github API is [here](https://developer.github.com/v3/).

````python
# List forks from url and get the info
forks_ih_url = "https://api.github.com/repos/ironhack-datalabs/bcn-feb-2019/forks"
forks = json.loads(gh_session.get(forks_ih_url).text)
forks
````

````python
languages = []  #create empty list to append all languages

for fork in forks:         #iterate throw forks information and get the language atribute
    languages.append(fork['language'])
    
list(set(languages))   #making sure that each language only appears once
````

````python
# List commits from repo 
commits_ih_url = "https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/commits"
params = {'since':'2021-05-24T00:00:00Z'}  #2weeks ago -> 24 May 2021

commits = json.loads(gh_session.get(commits_ih_url, params = params).text)   #get commits info
commits
````
